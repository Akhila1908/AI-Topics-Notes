```python
# Regular expressions - website(regex101)
import re
text="hlo my mobile number is 1234567890.bye 1234589342"
pattern='\d{10}'
matches=re.findall(pattern,text)
print(matches)


text1="hlo my mobile number is (123)-456-7890.Bye 1234589342"
pattern1='\d{10}|\(\d{3}\)-\d{3}-\d{4}'    # | or is used find 1 or more patterns 
matches1=re.findall(pattern1,text1)
print(matches1)


chat1="hlo my mail id number is abcA_08a@gmail.com."
pattern2='[a-zA-Z0-9_]*@[a-zA-Z0-9_]*\.[a-zA-Z]*'    #  * (star*) for the sequencw of charaters
matches2=re.findall(pattern2,chat1)  # .(dot) can be any  single character
print(matches2)

chat2="hlo my order number 478237512 having the issue."
pattern3='order[^\d]*(\d*)'  #  ^ makes sure does not contains
matches3=re.findall(pattern3,chat2)  
print(matches3)


chat3="hlo my age 23."
pattern4='age \d*'  #  ^ makes sure does not contains
matches4=re.findall(pattern4,chat3)  
print(matches4)


chat4="hlo my age 23."
pattern5='age (\d*)'   #  * (star*) for the sequence of charaters/numbers
matches5=re.findall(pattern5,chat4)  
print(matches5)


chat5="hlo my age 23."
pattern6='age \d'  
matches6=re.findall(pattern6,chat5)  
print(matches6)


chat6="hlo i born in 2006.in piduguralla"
pattern7='born.*'  
matches7=re.findall(pattern7,chat6)  
print(matches7)
```

```python
import spacy
nlp=spacy.load("en_core_web_sm")
doc=nlp("Eating bananas by the monkey is so sweet like sri")
base_token=nlp("mango")
for token in doc:
    print(token.text ,"<-->",base_token.text ,":" ,token.similarity(base_token)) # it shows how similar that the words in the doc with the base_token
```

```python
import spacy

nlp = spacy.load("en_core_web_md")  # or "en_core_web_lg"

king = nlp.vocab["king"].vector
man = nlp.vocab["man"].vector
woman = nlp.vocab["woman"].vector
queen = nlp.vocab["queen"].vector

result = king - man + woman

from sklearn.metrics.pairwise import cosine_similarity
similarity = cosine_similarity([result], [queen])
print(similarity)
```
