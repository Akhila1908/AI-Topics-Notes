```python
import torch 
from torch.utils.data import Dataset
import numpy as np
import os
import random
import matplotlib.pyplot
from PIL import Image # Pyhton Image Library
```

```python
class CustomDataset(Dataset):
    def __init__(self,dir):
        self.dir = dir
        self.images = os.listdir(dir)
    def __len__(self):
        return len(self.images)
    def __getitem__(self,index):
        image_path = os.path.join(self.dir,self.images[index])
        image = Image.open(image_path).convert("RGB")
        return image
        
```

# basic terms 

```python
* Loss Computation --> Loss tells how far the prediction is from the correct answer.
* The forward pass is when data goes forward through the network(into the model) to make a prediction.
* Always call zero_grad() before backward()

1. Forward pass      → Predict
2. Loss computation  → Measure error
3. Zero gradients    → Clear old memory
4. Backward pass     → Find mistakes
5. Optimizer step    → Fix mistakes
```

# optimizers

```python
 1. Optimizers in PyTorch are essential for updating the parameters like weights of a neural network 
 2. and to minimize the loss function during training.
 
3 . types :
                                       
torch.optim.SGD	     Stochastic Gradient Descent (with optional momentum)	--> model.parameters(), lr=0.01  
torch.optim.Adam	 Adaptive Moment Estimation (good default choice)	
torch.optim.AdamW	 Adam with decoupled weight decay (better for regularization) -->	Adam = Momentum + RMSprop ,model.parameters(), lr=0.01 , betas=(0.9, 0.999)
torch.optim.RMSprop	 RMSProp optimizer (good for RNNs)	                     -->  model.parameters(), lr=0.01 , alpha=0.99
torch.optim.Adagrad	 Adagrad optimizer (adapts learning rate per parameter)	  -->  model.parameters(), lr=0.01
torch.optim.ASGD	 Averaged SGD	lr, lambd, alpha, t0, weight_decay
torch.optim.LBFGS	 Quasi-Newton method (for small datasets)
                                       
```

```python
| Optimizer      | Speed     | Stability | Usage        |
| -------------- | --------- | --------- | ------------ |
| SGD            | Slow      | Low       | Rare         |
| SGD + Momentum | Medium    | Medium    | CV           |
| RMSprop        | Fast      | Good      | RNN          |
| Adam           | Very Fast | Excellent | Default      |
| AdamW          | Best      | Best      | Transformers |
```

```python
import torch
import torch.nn as nn
import torch.optim as optim # optim - optimizers

# Example model 
model = nn.Linear(10, 1)    # A linear layer with 10 inputs and 1 output

# Loss function
criterion = nn.MSELoss()     # Mean Squared Error loss function

# Optimizer (Adam in this case)
optimizer = optim.Adam(model.parameters(), lr=0.001)

# Dummy data
x = torch.randn(5, 10)   # batch of 5 samples, each with 10 features
y = torch.randn(5, 1)

# Training step
for epoch in range(3):
    optimizer.zero_grad()       # Reset gradients
    outputs = model(x)          # Forward pass
    loss = criterion(outputs, y) # Compute loss
    loss.backward()             # Backpropagation
    optimizer.step()            # Update parameters
    
    print(f"Epoch {epoch+1}, Loss: {loss.item():.4f}")
```

# Backward propagation

```python
# it is used to calculate loss

import pandas as pd
import numpy as np
import torch
from torch.utils.data import TensorDataset
from torch.utils.data import random_split
from torch import nn  # nn - neural network
from torch.nn import functional as f
from sklearn.model_selection import train_test_split
```

```python
class NeuralNetwork(nn.Module):
    def __init__(self):        # linear activation function
        super().__init__()
        self.layer1 = nn.Linear(4,8)   # input , output , and hidden layers
        self.layer2 = nn.Linear(8,9)
        self.layer3 = nn.Linear(9,3)

    def forward(self,x):
        x = torch.relu(self.layer1(x))
        x = torch.relu(self.layer2(x))
        x = self.layer3(x)
        return x
```

```python
torch.manual_seed(41)
```

```python
from sklearn.datasets import load_iris
model = NeuralNetwork()
iris = load_iris(as_frame=True)
df = iris.frame
print(df.shape)
```

```python
df.isnull().sum()
```

```python
df.target.value_counts()
```

```python
df["species"] = df["target"].map({0:"setosa",1:"versicolor",2:"virginica"})
```

```python
from sklearn.model_selection import train_test_split
import torch
import torch.nn as nn

# Features & labels
X = df.drop(columns=["target", "species"]).values
y = df["target"].values

# Train-test split
x_train, x_test, y_train, y_test = train_test_split( 
    X, y, train_size=0.75, random_state=100
)

# Convert numpy → torch tensors
x_train = torch.FloatTensor(x_train)
x_test = torch.FloatTensor(x_test)

y_train = torch.LongTensor(y_train)
y_test = torch.LongTensor(y_test)

# Loss & optimizer
criterion = nn.CrossEntropyLoss()  # crossEntropy loss — a standard loss function for classification problems.
optimizer = torch.optim.Adam(model.parameters(), lr=1e-3)  # lr - learning rate(η).

#The learning rate controls how much the model's parameters are updated during each optimization step.
#It is a scaling factor for the gradients computed during backpropagation.

# Forward pass
outputs = model(x_train)

# Loss calculation
loss = criterion(outputs, y_train)
loss
```

```python
losses = []
eopchs = 1000  # training the model 1000 times , More epochs → better learning (up to a limit)
for eopch in range(eopchs):
    outputs = model.forward(x_train)
    loss = criterion(outputs, y_train)
    optimizer.zero_grad()   #it is critical step in the training loop that resets the gradients of all model parameters before computing new ones
    loss.backward()  # performs backpropagation -- it computes the gradients of the loss with respect to all the model parameters that have requires_grad=True.
    optimizer.step() # updates the model parameters based on the gradients that were computed during backpropagation

    if(eopch + 1) % 10 == 0:
        print(f'Eopch[{eopch+1}],loss:{loss.item():.4f}')
```
